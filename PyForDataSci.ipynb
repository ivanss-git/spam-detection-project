```python
string = "I'm \\ very tired"
raw_string = r"I \n am \n very \n tired"
print(string.upper()) # output: I'M \ VERY TIRED
print(string.find("ver")) # output: 6
print(string.replace("very", "not")) # output: I'm \ not tired
print(string[0:6]) # output: I'm \ 
print(string[::2]) # output: Im\vr ie
print(len(string)) # output: 16
print("It is true.. {}".format(string)) # output: It is true.. I'm \ very tired
print("It is true.. %s" % (string)) # output: It is true.. I'm \ very tired
print(raw_string) # output: I \n am.. instead of newline
```

    * AttributeError: When an attribute or method is accessed on an object that doesnt posses that specific attribute or method (like trying to use sort on a string)

## 1D Numpy

- Operations: Usually faster than in regular python. 
- Vector + and -: first component is sum of first component of u and v, same for second components.
    * u: top of u is right, bottom is left
    * v: top of v is bottom of vector, bottom of v is top of vector
    * Tip to tail method 
    * the numpy code will run much faster and requires less code
- Array multiplication with a scalar:
    * only requires one line of code using numpy and is much faster
    * hadamard product: element-wise/point-wise product takes two vecgors of same dimension and multiplies their corresponding elements together to produce a new vector of the same size
    * basically the same operations from cal3/physics2 but only faster, more efficient, and with less code.
    * dot product implemented using .dot(u,v)
    * broadcasting: when adding a scalar to a np array, np adds it to each element automatically
    * 
- Universal functions operate on nd arrays. Ex: a = np.array([...]), then mean = a.mean(); others include .max() and more
    * we can also use uf's to map np arrays to new np arrays. 
    * .linspace(start,end,samples to generate) to plot mathematical functions; returns evenly space numbers over a specified interval

```python
import numpy as np
np.pi # pi in radians
x = np.array([0,np.pi/2, np.pi]) # x = [0, pi/2, pi]
y = np.sin(x) # y = [sin(0), sin(pi/2), sin(pi)]

# y: array([0,2,1.2e-16]) , y = [0,1,0]

np.linspace(-2,2,num=9)
# output: -2,-1.5,-1,-0.5,0,0.5,1,1.5,2

import matplotlib.pyplot as plt
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)
%matplotlib inline
plt.plot(x,y)
#output is a graph
```
