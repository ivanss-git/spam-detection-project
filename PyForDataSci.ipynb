## Operations in Python

* `/` — Performs standard floating-point division.

* `//` — Performs floor (integer) division.

* **Slicing:** Starts at the specified index and goes up to, but does not include, the ending index.
  Example:
  `variable[0:3]` returns the elements at indices `0`, `1`, and `2`.

* **Stride Slicing:** Selects every *n*th element in a sequence.
  Example:
  `name[::2]`

* **Tuples:** Immutable. `len()` returns the length of a tuple (or other sequence). Tuples can also be multiplied to repeat their contents.
  Example:
  `tuple[start:end]`

* **Concatenation:** Similar to Java, using the `+` operator.

* **String Escape Characters:**

  * `\n` — New line.
  * `\t` — Tab.
  * `\\` — Backslash.

* **String Methods:**

  * `.upper()` converts a string to uppercase.
  * `.lower()` converts a string to lowercase.
  * `.replace(old, new)` replaces one substring with another.

* `.find()` locates a substring and returns the starting index where it was found. If the substring is not found, it returns `-1`.

* **String Formatting:**

  * `"{}".format(value)` inserts values into a string.
  * `%s` formats a string.
  * `%d` formats an integer.

* **Raw Strings:** Prefix a string with `r`, such as `r"..."`, to treat backslashes as literal characters instead of escape sequences.



````python
string = "I'm \\ very tired"
raw_string = r"I \n am \n very \n tired"
print(string.upper()) # output: I'M \ VERY TIRED
print(string.find("ver")) # output: 6
print(string.replace("very", "not")) # output: I'm \ not tired
print(string[0:6]) # output: I'm \ 
print(string[::2]) # output: Im\vr ie
print(len(string)) # output: 16
print("It is true.. {}".format(string)) # output: It is true.. I'm \ very tired
print("It is true.. %s" % (string)) # output: It is true.. I'm \ very tired
print(raw_string) # output: I \n am.. instead of newline
````

## Exception Handling

* Exception handling is the process of dealing with errors so a program can continue running.

* **`try` / `except`:** Attempts to execute a block of code, then searches for the appropriate exception to handle any errors that occur.

* **Best practice:** Add an `else` statement to execute code only if no exceptions were raised, ensuring the program ran successfully.

* **Errors** are typically not caught or handled by the program itself, but by the environment or hardware. **Exceptions** can be caught and usually result from problematic code that may be expected.

  * **`ZeroDivisionError`**: Raised when dividing by zero.
  * **`ValueError`**: Raised when an inappropriate value is used (e.g., converting a non-numeric string to an integer with `int("abc")`).
  * **`FileNotFoundError`**: Raised when attempting to access a file that does not exist.
  * **`IndexError`**: Raised when attempting to access an element outside the valid index range (e.g., `my_list = [1, 2, 3]` and `my_list[5]`).
  * **`KeyError`**: Raised when attempting to access a key that does not exist in a dictionary.
  * **`TypeError`**: Raised when an object is used in an incompatible manner (e.g., `"String" + 5`).
  * **`AttributeError`**: Raised when an attribute or method is accessed on an object that does not have that attribute or method (e.g., calling `.sort()` on a string).
  * **`ImportError`**: Raised when attempting to import a module that is unavailable (e.g., `import non_existent_module`).

## Reading Files with `open()`

* Define a file using:
  `file = open("/directory/Examples.txt", "r")`
  The second parameter specifies the mode, such as reading (`"r"`), writing (`"w"`), or appending (`"a"`).

* **Attributes** allow us to get specific information about the file, such as `.name` and `.mode`.

* **Additional methods** include:

  * `.readline()` reads the next line of the file. Calling it repeatedly returns the next line each time. You can also use a loop to read all lines.
  * `.readlines()` returns all lines as a list.
  * `.read(#)` reads a specified number of characters from the current file position.

* Use `with open()` because it automatically closes the file when finished. The syntax is:
  `with open("filename.txt", "mode") as file:`

* You can verify whether the file is closed using:
  `print(file.closed)`

  Variables created inside the `with` block can still be used outside the block if they were defined beforehand.

* You can move to a specific position in the file using:
  `file.seek(#)`
  The position is specified in **bytes**, so you need to know the offset you want to read from.

* You can also read a specific number of characters using:
  `characters = file.read(5)`
  This reads 5 characters starting from the current position of the file pointer.

* As a best practice, always close a file when you are done using:
  `file.close()`
  (This is done automatically when using `with open()`.)

* **Writing to a file:** The same `open()` process is used, but with a write mode (`"w"` or `"a"`). Use:
  `file.write(data)`
  where the argument is the text you want to write to the file.

* **Copying one file to another:** Open the source file in read mode and the destination file in write mode using nested `with open()` statements, then use a loop to read from the source file and write each line to the destination file.

## 1D NumPy

* NumPy is a library for scientific computing with advantages such as speed and memory efficiency. It is also the foundation of pandas.

* A NumPy array is similar to a list, except it has a fixed size and typically stores elements of the same data type. You can cast a list to a NumPy array using `var_name = np.array([])` and access elements by index. The object type is `numpy.ndarray`, and the data type can be viewed with `a.dtype`, which may return something like `dtype('int64')` if all elements are of the same type.

  * `.size` returns the number of elements in the array.
  * `.ndim` returns the number of dimensions.
  * `.shape` returns a tuple representing the size of the array in each dimension.

* **Slicing and Indexing:** Similar to lists and tuples. We still do not include the element at the ending index.

* **Operations:** Usually much faster than regular Python.

* **Vector Addition and Subtraction:**

  * The first component is the sum (or difference) of the first components of vectors `u` and `v`, the second component is the sum (or difference) of the second components, and so on.
  * `u`: Top of `u` is the right side, and the bottom is the left side.
  * `v`: The top of `v` is the bottom of the vector, and the bottom of `v` is the top of the vector.
  * Tip-to-tail method.
  * NumPy performs these operations much faster and with less code.

  * Only requires one line of code using NumPy and is much faster.
  * **Hadamard Product:** An element-wise (point-wise) product that takes two vectors of the same dimension and multiplies their corresponding elements together to produce a new vector of the same size.

  * **Dot Product:** Implemented using `.dot(u, v)`.
  * **Broadcasting:** When adding a scalar to a NumPy array, NumPy automatically adds it to every element.

## HTTP Protocol

* **Uniform Resource Locator (URL):** A URL is the most common way to find resources on the web. It consists of a URL, request, and response.

* **URL Components** include:

  * **Scheme:** The protocol, such as `http://`
  * **Internet Address (Base URL):** For example, `www.ibm.com`
  * **Route:** The location of a resource on the web server, such as `/images/IDSNlogo.png`

  Combined, this becomes:

  `http://www.ibm.com/images/IDSNlogo.png`

* HTTP is the general protocol for transferring data across the web, including many types of REST APIs.

* When you use a webpage, the browser sends a request to the server. The server tries to find the requested resource, typically an `index.html` file by default. If successful, the server sends the object back to the client in an HTTP response, including information such as the content type, length, etc. Common resources stored on a web server include `index.html`, `image.png`, and `file.txt`.

* **Request and Response Process:**

  * **Request Start Line:** Contains the HTTP method, such as `GET /index.html HTTP/1.0`.
  * **Request Header:** Passes additional information with the HTTP request, such as `User-Agent: python-requests/2.21.0` and `Accept-Encoding: gzip, deflate`.
  * **Response Start Line:** Contains the HTTP version number, status code, and descriptive phrase, such as `HTTP/1.0 200 OK`.
  * **Response Header:** Contains information such as `Server: Apache` and `Cache: UNCACHEABLE`.
  * **Response Body:** Contains the requested file, which in this case could be an HTML document beginning with `<!DOCTYPE html>`.

* **Status Code Examples:**

  * **1XX:** Informational

    * `100` – Everything so far is OK.
  * **2XX:** Success

    * `200` – OK.
  * **3XX:** Redirection

    * `300` – Multiple Choices.
  * **4XX:** Client Error

    * `401` – Unauthorized.
    * `403` – Forbidden.
    * `404` – Not Found.
  * **5XX:** Server Error

    * `501` – Not Implemented.

* The first digit indicates the class of the status code:

  * **1XX** – Informational responses.
  * **2XX** – Successful responses.
  * **4XX** – Client errors ("bad news").
  * **5XX** – Server errors.

* **HTTP Methods** include:

  * **GET:** Retrieve data from the server.
  * **POST:** Submit data to the server.
  * **PUT:** Update data already on the server.
  * **DELETE:** Delete data from the server.

## REST APIs and HTTP Requests

* `requests` is a Python library that allows you to send HTTP/1.1 requests easily (`import requests`).

* You can make a request using the `get()` method. For example:


````python
import requests
url='https://www.ibm.com/'
r=requests.get(url)

# implement using attributes like .body, .headers, .encoding, .text
r.status_code :200 # requests have no body to r.request.nody returns none
header=r.headers # outputs a long body
````

* You can also use the GET method to modify the results.

* **GET Requests with URL Parameters:** The base URL is `httpbin.org`, and the route is `/get`, giving us `http://httpbin.org/get`. After the GET request, we have a query string that begins with `?`, followed by parameter name-value pairs in the form `parameter=value`, separated by `&`. In this example, the query string is `?name=Joseph&ID=123`. Combined, the full URL becomes:

`http://httpbin.org/get?name=Joseph&ID=123`


````python
# Python example for creating a query string

# Base URL with "get" appended to the end
url_get = "http://httpbin.org/get"

# Create a query string using a dictionary payload
payload = {"name": "Joseph", "ID": 123}

# Pass the dictionary payload to the params parameter of the GET request
r = requests.get(url_get, params=payload)

# Print the URL to see the query string with the names and values
r.url
# 'http://httpbin.org/get?name=Joseph&ID=123'

# Since the information is sent in the URL, the request body is empty
r.request.body
# None

# Check the status code
r.status_code
# 200

# View the response text
r.text

````

## Web Scraping

* The process of extracting data from websites or webpages. It involves automated retrieval from web sources and works through HTTP requests.

* Process:

  * **HTTP Request:** A web scraper sends an HTTP request to a URL, usually a GET request to retrieve content. The web server responds by returning the requested webpage's HTML content, which includes visible text, media elements, and the underlying HTML structure that defines the layout.
  * **HTML Parsing:** Once the HTML content is received, you need to break it down into components, or parse it, such as tags, attributes, and text content.
  * **Data Extraction:** From the parsed HTML content, web scrapers can identify and extract specific data, such as text, images, tables, prices, etc. Scrapers locate the data using relevant tags, attributes, and patterns in the HTML structure.
  * **Data Transformation:** Extracted data may need further processing and transformation, such as cleaning up the body and tags, to ensure it is ready for data analysis.
  * **Storage:** After extraction and transformation, we can store the data in databases, spreadsheets, JSON, or CSV files. The format choice depends on the requirements.
  * **Automation:** This entire process is automated in many cases. It is useful when extracting data from websites that regularly update their data.

* **HTML Structure:** Serves as the foundation of webpages. Examples include:

  * `<html>` for the root element of an HTML page.
  * `<head>` contains meta-information about the HTML page.
  * `<body>` displays the content.
  * `<h>` tags are used for different types of headings.
  * `<p>` is used for paragraphs and may contain information such as player salaries.

* HTML tags define the structure and consist of opening, or start, tags and closing, or end, tags. They have names and may contain attributes that provide additional information.

* **HTML Structure:** It is easier to visualize HTML documents as a tree, with tags as nodes. These nodes may contain strings, children, parents, siblings, and descendants.

* **Tables:** Tables present structured data using the `<table>` tag. Rows are defined using the `<tr>` tag, with the first row often being defined using `<th>` tags. Table cells are represented by `<td>` tags, which define individual cells within a row.

* The final web scraping step-by-step process requires two modules: `requests` and `BeautifulSoup`.


````python
import requests
from bs4 import BeautifulSoup

# Specify the URL of the webpage you want to scrape
url = 'https://en.wikipedia.org/wiki/IBM'

# Send an HTTP GET request to the webpage
response = requests.get(url)

# Store the HTML content in a variable
html_content = response.text

# Create a BeautifulSoup object to parse the HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Display a snippet of the HTML content
print(html_content[:500])
````

- Beautiful Soup allows you to find elements based on tags, attributes, and text, extracting only what you want.

- Using pandas `read_html` automatically extracts data from a website's tables and presents it in a suitable format. It's similar to taking a table from a webpage and importing it into a spreadsheet for further analysis.

- The correct scraping method depends on how the target website delivers data.
    * API Interception: Best for speed and structured data when hidden APIs are available.
    * HTML Parsing: Best for static sites with predictable layouts and no login walls.
    * Headless Browsers: Best for complex, dynamic sites that rely heavily on user interactions or JavaScript.
